```python
box_count = 0

for ele in unique_elements:
    x_inds = np.where(labels==ele)[0]
    y_inds = np.where(labels==ele)[1]

    unique_x_inds = np.unique(x_inds)
    unique_y_inds = np.unique(y_inds)

    unique_x_quotient_inds = (unique_x_inds)//64
    unique_unique_x_quotient_inds = np.unique(unique_x_quotient_inds)


    unique_y_quotient_inds = (unique_y_inds)//64
    unique_unique_y_quotient_inds = np.unique(unique_y_quotient_inds)

    box_bottom_left_x = unique_unique_x_quotient_inds*64
    box_bottom_left_y = unique_unique_y_quotient_inds*64


    print("box_bottom_left_x, box_bottom_left_y", box_bottom_left_x, box_bottom_left_y)

    print("len(box_bottom_left_x), len(box_bottom_left_y)", len(box_bottom_left_x), len(box_bottom_left_y))

    box_count += len(box_bottom_left_x)*len(box_bottom_left_y)
print("box_count", box_count)
```
